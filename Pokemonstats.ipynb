```python
def get_stats(name, url):
    
    response = rq.request("GET", url)
    data = json.loads(response.text)
    stats = data['stats']
    
    df_stats = pd.DataFrame(index=[name])

    n=0
   
    Statsloop = True

    while Statsloop:
        try:
            stat = stats[n]['stat']['name']
            value = stats[n]['base_stat']
            df_stats[stat] = [value]
            print(stat)
            print(value)
        except:
            Statsloop = False
        n += 1
    return df_stats
    
```

```python
name = 'bulbasaur'
url = 'https://pokeapi.co/api/v2/pokemon/1/'

get_stats(name,url)
```

## Batch data

```python
name = 'bulbasaur'
url = 'https://pokeapi.co/api/v2/pokemon/'
response = rq.get(url)
data = json.loads(response.text)

results = data['results']
results[0]['url']
```

```python
def get_batch(results):
    batch_stat_df = pd.DataFrame()

    for result in results:
        name = result['name']
        url = result['url']
        #print(get_stats(name,url))
        #print(url)
        #print(get_stats_types(name,url))
        batch_stat_df = pd.concat([batch_stat_df,get_stats(name,url)])
    return batch_stat_df
```

```python
data['results']
```

```python
url = 'https://pokeapi.co/api/v2/pokemon/'

response = rq.request("GET", url)
data = json.loads(response.text)
get_batch(data['results'])
```

```python
url = 'https://pokeapi.co/api/v2/pokemon/'
next_url = url
pokemon_df = pd.DataFrame()
stop = False
while not stop:
    try:
        response = rq.request("GET", next_url)
        data = json.loads(response.text)
    except:
        stop = True
        break
    pokemon_df = pd.concat([pokemon_df,get_batch(data['results'])])
    next_url = data['next']
    print(next_url)
```

```python
pokemon_df
```

```python
pokemon_df = pokemon_df.rename(index={'': "Name"})
```
